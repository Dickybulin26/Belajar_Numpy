# **MEMVISUALISASIKAN DATA DENGAN MATPLOTLIB DAN NUMPY**

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# PERSAMAAN GARIS
## Y = 2X + 3

```python
x = np.arange(1,11,1)
y = 2 * x + 3
print(x)
print(y)
```

```python
plt.figure(1)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

# LINGKARAN

```python
jari2 = 5
sudut = np.linspace(0,2*np.pi,100)
# sudut = np.arange(0,2*np.pi,0.1)

x2 = jari2*np.cos(sudut)
y2 = jari2*np.sin(sudut)
```

```python
plt.figure(2)
plt.plot(x2,y2)
plt.grid(True)
plt.show()
```
